```python
from sklearn import datasets
iris=datasets.load_iris()
df=pd.read_csv(iris.filename)
df.describe()
```

```python
df_with_columns=pd.read_csv(r'C:\\Users\\NODE\\anaconda3\\lib\\site-packages\\sklearn\\datasets\\data\\iris.csv',
                           delimiter=',',
                           header=0,
                           names=['sepal length (cm)','sepal width (cm)',
                                 'petal length (cm)','petal width (cm)','Variety']
                           )
df_with_columns.head()
```

```python
print(df_with_columns.shape)
df_with_columns["Variety"].hist()
```

```python
print(df_with_columns.shape)
print(df_with_columns.columns)
```

```python
print(df_with_columns.describe())
```

```python
for ohja,feature in enumerate(list(df_with_columns.columns)[:-1]):
    fg=sns.FacetGrid(df_with_columns,hue='Variety',height=5)
    fg.map(sns.displot,feature).add_legend()
    plt.show()
```

```python
sns.boxplot(x='Variety',y='sepal width(cm)', data=df_with_columns)
plt.show()
```

```python
sns.violinplot(x='Variety',y='sepal length(cm)', data=df_with_columns, size=20)

plt.show()
```

```python
df_with_columns.plot(kind='scatter',x='sepal length(cm)',y='sepal width(cm)', c= 'g')

plt.show()
```

```python
sns.set_style("whitegrid");

sns.pairplot(df_with_columns,vars = iris.feature_names, hue = 'Variety', diag_kind = 'kde', 

       plot_kws = {'alpha': 0.6, 's': 80, 'edgecolor': 'k'}, height = 2) 

plt.show()
```
